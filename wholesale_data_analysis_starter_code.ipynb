## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']

df[['client_id', 'unit_price', 'qty', 'subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

df['total_weight']= df['unit_weight'] * df['qty']

for i in range(len(df)):
    if df.loc[i, 'total_weight'] > 50:
        df.loc[i, 'shipping_price'] = 7 * df.loc[i, 'total_weight']
    else:
        df.loc[i, 'shipping_price'] = 10 * df.loc[i, 'total_weight']


df[['client_id', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = ((df['subtotal'] + df['shipping_price'])*1.0925).round(2)

df[['client_id', 'subtotal', 'shipping_price', 'total_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head(10)
```

```python
# Create a column for the profit of each line using line cost and line price

df['line_profit'] = df['total_price'] - df['line_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
#show the top 5 clients_ids by quantity and how much the spent

top_five_client_entries = df['client_id'].value_counts().head(5).index.tolist()

#loop through the top 5 clients and calculate the total price
for client in top_five_client_entries:
    total_price = df[df['client_id'] == client]['total_price'].sum()
    print(f"Client {client} spent ${round(total_price, 2)}")    
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_five_client_entries = df['client_id'].value_counts().head(5).index.tolist()


# Filter the DataFrame to include only the top 5 clients
top_clients_df = df[df['client_id'].isin(top_five_client_entries)]
# top_clients_df.head()

# Group by client_id and aggregate the required columns
top_clients_summary_df = top_clients_df.groupby('client_id').agg(
    total_units=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('line_profit', 'sum')
).reset_index()

top_clients_summary_df.head()
```

```python
top_clients_summary_df.columns
```

```python
# Format the data and rename the columns to names suitable for presentation.
renamed_columns_df = top_clients_summary_df.rename(columns={
    'client_id': 'Client ID',
    'total_units': 'Units',
    'total_shipping_price': 'Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_profit': 'Total Profit (millions)'
})

# renamed_columns_df.columns

# Define the money columns.
money_columns = renamed_columns_df.columns.str.contains('millions') 

#print the money columns
money_columns

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return f"${value/1000000:.2f}"

#  Apply the currency_format_millions function to only the money columns. 
for column in renamed_columns_df.columns[money_columns]:
    renamed_columns_df[column] = renamed_columns_df[column].apply(currency_format_millions)

renamed_columns_df.head()


# Rename the columns to reflect the change in the money format. 
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df = renamed_columns_df.sort_values('Total Profit (millions)', ascending=False)
sorted_df.head()
```

```python
# calculate the sum of the total revenue for the top 5 clients

total_revenue = sorted_df['Total Revenue (millions)'].apply(lambda x: float(x[1:])).sum()
print(f"Total revenue for the top 5 clients: ${total_revenue}")
total_profit = sorted_df['Total Profit (millions)'].apply(lambda x: float(x[1:])).sum()
print(f"Total profit for the top 5 clients: ${total_profit}")
```
